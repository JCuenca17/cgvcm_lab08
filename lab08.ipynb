```python
import cv2
img = cv2.imread('gato.jpg')
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
centro = (210, 130)
radio = 80
color = (255, 0, 0)
cv2.circle(fix_img, centro, radio, color, thickness=2)
plt.imshow(fix_img)
```

```python
texto = "gato"
posicion_texto = (180, 250)
font = cv2.FONT_HERSHEY_SIMPLEX
tamaño_letra = 1
color_letra = (255, 255, 255)
grosor_letra = 3
cv2.putText(fix_img, texto, posicion_texto, font, tamaño_letra, color_letra, grosor_letra)
plt.imshow(fix_img)
```

```python
new_img = cv2.cvtColor(fix_img,cv2.COLOR_RGB2BGR)
plt.imshow(new_img)
```

```python
cv2.imwrite('gato_nuevo.jpg',new_img)
```

```python
ret,thresh1 = cv2.threshold(fix_img,100,255,cv2.THRESH_BINARY)
plt.imshow(thresh1,cmap='gray')
```
